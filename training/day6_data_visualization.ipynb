Task 1: Matplotlib Basics
• Import Matplotlib:
▪ import matplotlib.pyplot as plt
• Create:
o Line plot (plt.plot)
o Scatter plot (plt.scatter)
• Save a figure using:
• plt.savefig('figure1.png', dpi=300)

```python
df=pd.read_csv("E:/QuantLake/Iris.csv")
```

```python
df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Id'][:50], df['SepalLengthCm'][:50], color='blue', label='Sepal Length')
plt.xlabel('Sample Id')
plt.ylabel('Sepal Length (cm)')
plt.title('Line Plot of Sepal Length (First 50 Samples)')
plt.legend()
plt.tight_layout()
plt.savefig('line_plot_iris.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], color='green', alpha=0.6)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.tight_layout()
plt.savefig('scatter_plot_iris.png', dpi=300)
plt.show()
```

Task 2: Plot Customization
• Add:
o Titles
o X/Y-axis labels
o Legends
• Use plt.grid(), change figure size, and set axis limits if needed

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df['Id'][:50], df['SepalLengthCm'][:50], color='darkblue', linestyle='-', marker='o', label='Sepal Length')
plt.title('Sepal Length of First 50 Iris Samples', fontsize=14, weight='bold')
plt.xlabel('Sample ID')
plt.ylabel('Sepal Length (cm)')
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend(loc='upper left')
plt.xlim(0, 55)  # optional
plt.ylim(df['SepalLengthCm'][:50].min() - 0.5, df['SepalLengthCm'][:50].max() + 0.5)
plt.tight_layout()
plt.savefig('customized_line_plot_iris.png', dpi=300)
plt.show()
```

Task 3: Seaborn Introduction
• Import Seaborn:
o import seaborn as sns
• Create:
o Boxplot (sns.boxplot)
o Heatmap (sns.heatmap)
o Histogram (sns.histplot)
• Try visualizing correlation or outliers

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=df, palette='Set2')
plt.title('Boxplot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('boxplot_sepal_length_species.png', dpi=300)
plt.show()
```

```python
# Compute correlation matrix
corr = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig('heatmap_iris.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='PetalLengthCm', bins=30, kde=True, color='skyblue')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('histogram_petal_length.png', dpi=300)
plt.show()
```

Task 5: Produce Publish-Quality Figures
• Refine:
o Chart DPI
o Colors
o Fonts and layout
• Matplotlib is designed for "publication-quality figures" — polish one chart to that
level

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style and scaling for publication
sns.set_theme(style='whitegrid', font_scale=1.3)

# Create high-quality figure
plt.figure(figsize=(10, 6), dpi=300)
box = sns.boxplot(
    data=df,
    x='Species',
    y='PetalWidthCm',
    palette='Set2',
    linewidth=1.5,
    fliersize=3  # small dots for outliers
)

# Title and labels
plt.title('Petal Width Distribution by Iris Species', fontsize=16, weight='bold')
plt.xlabel('Species', fontsize=13)
plt.ylabel('Petal Width (cm)', fontsize=13)

# Grid and layout refinements
plt.grid(True, axis='y', linestyle='--', linewidth=0.6, alpha=0.7)
sns.despine()  # Remove top and right borders
plt.tight_layout()

# Save the polished figure
plt.savefig('iris_boxplot_published.png', dpi=300)
plt.show()
```
