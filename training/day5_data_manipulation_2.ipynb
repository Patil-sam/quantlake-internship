```python
Task 1: Reshaping DataFrames
• Learn and practice:
o df.pivot(index, columns, values)
o df.pivot_table(values, index, columns, aggfunc)
o df.melt() to unpivot wide to long format
• Examples:
o Pivot total sales by Region and Category
o Use pivot_table to get average profit per Segment
```

```python
import pandas as pd
```

```python
df=pd.read_csv("Super_Store_Sales_data.csv")
```

```python
df
```

```python
pivot_sales=df.pivot_table(index="Region",columns="Category",values="Sales",aggfunc="sum")
```

```python
pivot_sales
```

```python
pivot_table_avg=df.pivot_table(values="Sales",index="Segment",aggfunc="mean")
```

```python
pivot_table_avg
```

```python
melted_df = df.melt(
    id_vars=['Order ID', 'Region', 'Category'],
    value_vars=['Sales'],
    var_name='Metric',
    value_name='Value'
)
```

```python
melted_df
```

```python
Task 2: Apply Custom Functions
• Use .apply() with custom lambda functions on columns
• Create a column that:
o Classifies profit margins (e.g. High/Medium/Low)
o Flags rows based on a condition (e.g. > 90% discount)
```

```python
df['Sales_Margin'] = df['Sales'].apply(
    lambda x: 'High' if x > 1000 else ('Medium' if x >= 500 else 'Low')
)

df[['Sales', 'Sales_Margin']]
```

```python
if 'Sales' in df.columns:
    df['High_Sales_Flag'] = df['Sales'].apply(lambda x: 'Yes' if x > 300 else 'No')
    print(df[['Sales', 'High_Sales_Flag']].head(10))
else:
    print("Column 'Sales' not found in dataset.")
```

```python
Task 3: Mapping & Replacing
• Use .map() and .replace() to clean or transform column values
• Example:
o Map full country names to codes
o Replace "Consumer" with "Retail" in Segment column
```

```python
df["Country"].unique()
```

```python
# Example mapping full country names to 2-letter codes
country_map = {
    'United States': 'US',
}

df['Country_Code'] = df['Country'].map(country_map)

df[['Country', 'Country_Code']].drop_duplicates()
```

```python
df['Segment'] = df['Segment'].replace('Consumer', 'Retail')

# Preview the change
df['Segment'].value_counts()
```

```python
df['Segment'] = df['Segment'].replace({
    'Consumer': 'Retail',
    'Corporate': 'Business',
    "Home Office" : "home"
})

df['Segment'].value_counts()
```

```python
Task 4: Combining DataFrames
• Concatenate vertically and horizontally using:
o pd.concat([df1, df2])
o Use ignore_index=True, axis=0/1
• Try combining two DataFrames with overlapping or mismatched columns
```

```python
df1 = pd.DataFrame({
    'Order ID': ['CA-1001', 'CA-1002'],
    'Customer Name': ['Alice', 'Bob'],
    'Sales': [250, 300]
})
```

```python
df2 = pd.DataFrame({
    'Order ID': ['CA-1003', 'CA-1004'],
    'Customer Name': ['Charlie', 'David'],
    'Sales': [150, 500]
})
```

```python
combined_df = pd.concat([df1, df2], ignore_index=True)
combined_df
#Vertical Concatenation (axis=0)
```

```python
#Horizontal Concatenation (axis=1)
horizontal_df = pd.concat([df1, df2], axis=1)
horizontal_df
```

```python
df3 = pd.DataFrame({
    'Order ID': ['CA-1005'],
    'Region': ['West'],
    'Profit': [100]
})
```

```python
mismatch_df = pd.concat([df1, df3], ignore_index=True)
print("Vertical with mismatched columns:\n", mismatch_df)
```
