Task 4: GroupBy Operations
• Use df.groupby('Category').mean()
• Try multiple aggregations:
• df.groupby('Region')['Sales'].agg(['sum', 'mean', 'count'])
• Apply reset_index() to clean up grouped results

```python
# Average of numeric columns grouped by Category
category_mean = df.groupby('Category').mean(numeric_only=True)
category_mean
```

```python
region_sales_agg = df.groupby('Region')['Sales'].agg(['sum', 'mean', 'count'])
region_sales_agg
```

```python
# Aggregate sales by Region
region_sales_agg = region_sales_agg.reset_index()
region_sales_agg
```

```python
 # Reset index to make 'Region' a regular column
subcat_stats = df.groupby(['Category', 'Sub-Category'])['Sales'].agg(['sum', 'mean', 'count']).reset_index()
subcat_stats
```

```python
# Group by Category & Sub-Category with multiple aggregations
segment_region = df.groupby(['Segment', 'Region'])['Sales'].agg(['sum', 'mean']).reset_index()
segment_region
```

Task 5: Merging DataFrames
• Create or use two small tables (e.g. orders.csv + customers.csv)
• Merge using:
o pd.merge(left, right, on='key')
o Join types: inner, outer, left

```python
#Orders Data
orders = pd.DataFrame({
    'Order ID': ['CA-2016-152156', 'CA-2016-152157', 'US-2017-108966'],
    'Customer ID': ['CG-12520', 'DV-13045', 'SO-20335'],
    'Sales': [261.96, 731.94, 14.62]
})
```

```python
orders
```

```python
#Customers Data
customers = pd.DataFrame({
    'Customer ID': ['CG-12520', 'DV-13045', 'TA-21050'],
    'Customer Name': ['Claire Gute', 'Darrin Van Huff', 'Tamara Chand'],
    'Segment': ['Consumer', 'Corporate', 'Home Office']
})
```

```python
customers
```

```python
# Merge with only matching Customer IDs
inner_merged = pd.merge(orders, customers, on='Customer ID', how='inner')
inner_merged
```

```python
 # Merge all orders and match customer info where available
left_merged = pd.merge(orders, customers, on='Customer ID', how='left')
left_merged
```

```python
# Merge all data from both orders and customers
outer_merged = pd.merge(orders, customers, on='Customer ID', how='outer')
outer_merged
```
